## Problem description

Air pollulants tend to accumulate in the lowest layer of the atmosphere known as the troposphere. This accumulation is mainly due to the phenomenon of thermal inversion. During thermal inversion, a layer of warm air moves over a layer of cold air, causing a disruption in the airflow. The layer of cold air is trapped between earth's surface and the warm air layer above it. the trapped air cannot carry the pollulants away and they begin to accumulate over time. 

Cities are particularly affected by thermal inversion given their constant production of pollulants during transportation and industrial activities. Additionally, the topography of city can worse the effect. For example, in cities surrounded by mountains, it is easier for cold air to become trapped during thermal inversion. Such is the case of Los Angeles, Mexico City and others.

### E. Assuming that the volumetric flow rates entering and leaving the basin are identical $\nu = \nu_0$. Show that the unsteady mole balance on CO within the basin becomes: $$F_{CO,A} + F_{CO,S} - v_OC_{CO} = V(\frac{dC_{CO}}{dt})$$

#### General Mole Balance

#### In - out + Generation = Accumulation
**In:** Moles of CO entering the basin = Moles entering from the Santa Ana wind = $F_{CO,S}$ <br>
**Generation:** Moles of CO produced by the cars =  $F_{CO,A}$ <br>
**Out:** $F_{CO}$ <br>
**Accumulation:** $ \frac {d(N_{CO})} {dt} $ <br>

$$F_{CO,S} - F_{CO} + F_{CO,A} = \frac {d(N_{CO})} {dt} $$

Since $\nu = \nu_0$, then $F_{CO} = \nu*C_{CO} = \nu_{0}*C_{CO} $

On the other hand, $N_{CO} = V*C_{CO}$

Applying these changes to the previous equation:

$$F_{CO,S} - \nu_{0}C_{CO} + F_{CO,A} = \frac {d(VC_{CO})} {dt} $$

Assuming the system volume is constant across time, then we obtain the final equation:

$$F_{CO,S} - \nu_{0}C_{CO} + F_{CO,A} = V \frac {d(C_{CO})} {dt} $$


### F. Verify that the solution to the previous equation is $t = \frac{V}{\nu_0}ln(\frac{[F_{CO,A}+F_{CO,S}]-\nu_0C_{CO,O}}{[F_{CO,A}+F_{CO,S}]-\nu_0C_{CO,f}})$

Beginning with the previous equation:

$$F_{CO,S} - \nu_{0}C_{CO} + F_{CO,A} = V \frac {d(C_{CO})} {dt} $$

assuming both $F_{CO,S}$ and $F_{CO,A}$ are both constant over time, we can combine both flowrates as:

$$ F_{CO,0} = F_{CO,S} + F_{CO,A} $$

$$F_{CO,0} - \nu_{0}C_{CO} = V \frac {d(C_{CO})} {dt} $$

$$\int_{0}^{t} dt = \int_{C_{CO,0}}^{C_{CO,f}} \frac {V} {F_{CO,0} - \nu_{0}C_{CO}} d(C_{CO}) $$

$$ t = \int_{C_{CO,0}}^{C_{CO,f}} \frac {V} {F_{CO,0} - \nu_{0}C_{CO}} d(C_{CO}) $$

$$ t = V \int_{C_{CO,0}}^{C_{CO,f}} \frac {1} {F_{CO,0} - \nu_{0}C_{CO}} d(C_{CO}) $$

Substituting: $U = F_{CO,0} - \nu_{0}C_{CO} $,  $dU = -\nu_{0}d(C_{CO}) $ <br>

$$ t = \frac {-V} {\nu_{0}} \int_{U_{CO,0}}^{U_{CO,f}} \frac {1} {U} dU $$

$$ t = \frac {-V} {\nu_{0}} ln (U) \Big|_{U_{CO,0}}^{U_{CO,f}} $$

Replacing the substitution

$$ t = \frac {-V} {\nu_{0}} ln (F_{CO,0} - \nu_{0}C_{CO}) \Big|_{C_{CO,0}}^{C_{CO,f}} $$

$$ t = \frac {-V} {\nu_{0}} \frac {ln (F_{CO,0} - \nu_{0}C_{CO,f})} {ln (F_{CO,0} - \nu_{0}C_{CO,0})}  $$

$$ t = \frac {V} {\nu_{0}} \frac {ln (F_{CO,0} - \nu_{0}C_{CO,0})} {ln (F_{CO,0} - \nu_{0}C_{CO,f})}  $$

$$ t = \frac {V} {\nu_{0}𝑡} \frac {ln ([F_{CO,S} + F_{CO,A}] - \nu_{0}C_{CO,0})} {ln ([F_{CO,S} + F_{CO,A}] - \nu_{0}C_{CO,f})}  $$

### H. There is heavier traffic in the L.A. basin in the mornings and in the evenings as workers go to and from work in downtown L.A. Consequently, the flow of CO into the L.A. basin might be better represented by the sine function over a 24-hour period. $$F_{CO,A} = a + b*sin(\frac{\pi t}{6}) $$

### the time, t = 0, starts at midnight and a = 35,000 $\frac{lb\ mol}{h}$ and b = 30,000 $\frac{lb\ mol}{h}$. The general material balance equation now becomes: $$a + b*sin(\frac{\pi t}{6}) + F_{CO,S} - v_OC_{CO} = V(\frac{dC_{CO}}{dt}) $$.

```python
F_sin_A = []
range_t = range(0,72,1)

a = 35000
b = 30000
    
for t in range_t:
    F_sin_A.append(a + b*np.sin(np.pi*t/6))
    
plt.plot(range_t, F_sin_A)
```

```python
def dCco_dt(C, t, a, b, F_CO_S, v_0, V):
    return [(a + b*np.sin(np.pi*t/6) + F_CO_S - v_0*C[0])/V]
```

```python
from scipy.integrate import odeint

ts = np.linspace(0,96,1000)
Cc0 = C_co_0  # the initial condition
C_CO = odeint(func=dCco_dt, y0=C_co_0, t=ts, args=(a, b, F_CO_S, v_0, V), rtol=1e-10, atol=1e-10)
ys = np.array(C_CO).flatten()

plt.xlabel("time (h)")
plt.ylabel("CO Concentration (lb mol/ft3)")
plt.plot(ts, ys);
```

#### Analytical Solution

```python
beta = np.pi/6
alpha = v_0/V

int_C = C_co_0 - (a+F_CO_S)/(v_0) - (b/V)*(beta/(beta**2 + alpha**2))

Cco = []
for t in ts:
    
    sec_1 = (a+F_CO_S)/(v_0)
    sec_2 = b*alpha*np.sin(beta*t)/(V*((beta**2 + alpha**2)))
    sec_3 = b*beta*np.cos(beta*t)/(V*((beta**2 + alpha**2)))
    sec_4 = int_C*np.exp(-alpha*t)
    
    Cco.append((sec_1 + sec_2 - sec_3 + sec_4))
    
plt.xlabel("time (h)")
plt.ylabel("CO Concentration (lb mol/ft3)")
plt.plot(ts, Cco)
```

### Appendix: Derivation of the analytical solution for differential equation on point H.

Original Differential Equation

$$a + b*\sin(\frac{\pi t}{6}) + F_{CO,S} - v_OC_{CO} = V(\frac{dC_{CO}}{dt}) $$.

Rearranging terms:

$$ \frac{dC_{CO}}{dt} + \frac {v_O}{V} C_{CO} =  \frac {a + b*\sin(\frac{\pi t}{6}) + F_{CO,S}} {V} $$.

This is a first-order differential equation of the form:

$$ \frac{dy}{dx} + P(x)y = Q(x) $$.

Hence, it can be solved via integrating factor with: $I = e^{\int_{}{} P(x)dt} = e^{\int_{}{} \frac {\nu_0}{V} dt} = e^{\frac {\nu_0}{V} t}$

$$ C_{CO} = \frac {\int_{}{} I Q(x)dx}{I} = \frac {\int_{}{} e^{\frac {\nu_0}{V} t} (\frac {a + b*\sin(\frac{\pi t}{6}) + F_{CO,S}} {V}) dt}{e^{\frac {\nu_0}{V} t}} $$

Resolving the integral in the numerator:

$$\int_{}{} e^{\frac {\nu_0}{V} t} (\frac {a + b*sin(\frac{\pi t}{6}) + F_{CO,S}} {V}) dt$$

$$\int_{}{} e^{\frac {\nu_0}{V} t} \frac {a}{V} dt + \int_{}{} e^{\frac {\nu_0}{V} t} \frac {F_{CO,S}}{V} dt + \int_{}{} e^{\frac {\nu_0}{V} t}\sin(\frac{\pi t}{6}) \frac {b}{V} dt$$

The solution of the first integral is:

$$\int_{}{} e^{\frac {\nu_0}{V}t} \frac {a}{V} dt = \frac {a}{V} \frac {V}{\nu_0} e^{\frac {\nu_0}{V}t} + C = \frac {a}{\nu_0} e^{\frac {\nu_0}{V}t} + C$$ 

The solution of the second integral is:

$$\int_{}{} e^{\frac {\nu_0}{V}t} \frac {F_{CO,S}}{V} dt = \frac {F_{CO,S}}{V} \frac {V}{\nu_0} e^{\frac {\nu_0}{V}t} + C = \frac {F_{CO,S}}{\nu_0} e^{\frac {\nu_0}{V}t} + C$$ 

The third integral is solved by integration by parts, the solution is:

$$\int_{}{} e^{\frac {\nu_0}{V} t}\sin(\frac{\pi t}{6}) \frac {b}{V} dt = $$ <br>
$$\frac {b}{V} e^{\frac {\nu_0}{V} t} \frac {1}{(\frac {\pi}{6})^2 + (\frac {v_0}{V})^2}(\frac {v_0}{V} \sin(\frac {\pi}{6}t) - \frac {\pi}{6} \cos(\frac {\pi}{6}t)) + C$$ 

Therefore:

$$ C_{CO} = \frac {a}{\nu_0} + \frac {F_{CO,S}}{\nu_0} + \frac {b}{V} \frac {1}{(\frac {\pi}{6})^2 + (\frac {v_0}{V})^2}(\frac {v_0}{V} \sin(\frac {\pi}{6}t) - \frac {\pi}{6} \cos(\frac {\pi}{6}t)) + Ce^{\frac {-\nu_0}{V}t}$$

$$ \nu_0 C_{CO} = a + F_{CO,S} + \frac {b}{V} \frac {\nu_0}{(\frac {\pi}{6})^2 + (\frac {v_0}{V})^2}(\frac {v_0}{V} \sin(\frac {\pi}{6}t) - \frac {\pi}{6} \cos(\frac {\pi}{6}t)) + \nu_0Ce^{\frac {-\nu_0}{V}t}$$ <br>

at t = 0, $C_{CO} = C_{CO}^{0}$ <br>
$$ \nu_0  C_{CO}^{0} = a + F_{CO,S} + \frac {b\nu_0}{V} \frac{1}{({(\frac {\pi}{6})^2 + (\frac {v_0}{V})^2})}(\frac {v_0}{V} \sin(\frac {\pi}{6}0) - \frac {\pi}{6} \cos(\frac {\pi}{6}0)) + \nu_0Ce^{\frac {-\nu_0}{V}0}$$

$$ C_{CO}^{0} = \frac {a}{\nu_0} + \frac {F_{CO,S}}{\nu_0} - \frac {b}{V} \frac {1} {{(\frac {\pi}{6})^2 + (\frac {v_0}{V})^2}}(\frac {\pi}{6}) + C$$

$$ C = C_{CO}^{0} - \frac {a}{\nu_0} - \frac {F_{CO,S}}{\nu_0} + \frac {b}{V} \frac {1} {{(\frac {\pi}{6})^2 + (\frac {v_0}{V})^2}}(\frac {\pi}{6}) $$
